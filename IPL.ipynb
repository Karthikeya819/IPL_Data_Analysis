## 16) Find the Total runs in each season

```python
seasons = sorted(set(df['season']))
seasons_group = df.groupby('season')
deliveries_group = df1.groupby('match_id')
for season in seasons:
    runs=0
    season_group = seasons_group.get_group(season)['id']
    for match in season_group:
        runs+=deliveries_group.get_group(match)['total_runs'].sum()
        runs+=deliveries_group.get_group(match)['extra_runs'].sum()
    print(str(season)+' - '+str(runs))
```

## 18) Visualize the Toss decision across seasons

```python
seasons = sorted(set(df['season']))
x = range(len(seasons))
seasons_group = df[['toss_decision','season']].groupby('season')
group2_data = [seasons_group.get_group(seasons[i])['toss_decision'].count() for i in x]
group1_data = [(seasons_group.get_group(seasons[i])['toss_decision']=='field').sum() for i in x]
group3_data = [(~(seasons_group.get_group(seasons[i])['toss_decision']=='field')).sum() for i in x]
bar_width = 0.3

plt.bar(x, group1_data, width=bar_width, label='Field', align='center')
plt.bar([i+ bar_width for i in x], group2_data, width=bar_width, label='Total Tosses', align='center')
plt.bar([i + 2 * bar_width for i in x], group3_data, width=bar_width, label='Bat', align='center')

plt.xlabel('Seasons')
plt.ylabel('Noof')
plt.title('Toss Decision Across Seasons')
plt.xticks([i+ bar_width for i in x],seasons)
plt.legend()
plt.show()
```

## 19) Find the Dismissal Kind and Visualize using best fit graph

```python
dismissal = df1['dismissal_kind'].dropna().unique()
x = range(len(dismissal))
dismissal_group = df1.groupby('dismissal_kind')
data = [dismissal_group.get_group(dismissal[i])['dismissal_kind'].count() for i in x]
bar_width = 0.4  
plt.bar([i for i in x], data, width=bar_width, align='center') 

for i in x:
    plt.text(i,data[i],str(data[i]), ha='center', va='bottom')
    
plt.xlabel('Dismissal Kind')
plt.ylabel('Frequency')
plt.title('Dismissal Kind in IPL')
plt.xticks([i for i in x], dismissal, rotation=45, ha='right')
plt.show()

```

## 20) Find the Top 10 run scorers in IPL and Visualize using best fit graph

```python
batsman_group = df1.groupby('batsman').sum()['total_runs'].sort_values(ascending=False).head(10)
batsmans = batsman_group.index

x = range(len(batsmans))
data = [batsman_group.iloc[i] for i in x]
bar_width = 0.4  

for i in x:
    plt.text(i,data[i],str(data[i]),ha='center',va='bottom')

plt.bar([i for i in x], data, width=bar_width, align='center') 

plt.xlabel('Crickecter')
plt.ylabel('Frequency')
plt.title('Dismissal Kind in IPL')
plt.xticks([i for i in x], batsmans, rotation=45, ha='right')
plt.show()
```

## 23) Compare Total Played Matches vs Winning Matches vs Win Rate

```python
Total_Matches_Counts = pd.concat([df['team1'],df['team2']]).value_counts()
Total_Matches = Total_Matches_Counts.index
Win_Matches_Count = df['winner'].value_counts()
data_group1=[] 
data_group2=[] 
teams=[]
for team in Total_Matches:
    teams.append(team)
    data_group1.append(Total_Matches_Counts.loc[team])
    data_group2.append(Win_Matches_Count.loc[team])

bar_width = 0.4
x = range(len(teams))

plt.bar(x, data_group1, width=bar_width, label='Total Played Matches', align='center')
plt.bar([i + bar_width for i in x], data_group2, width=bar_width, label='Winning Matches', align='center')

plt.xlabel('Teams')
plt.ylabel('Frequency')
plt.title('Total Played Matches vs Winning Matches')
plt.xticks([i for i in x], teams, rotation=45, ha='right')
plt.legend()
plt.show()
```

## 24) Find the Distribution of Won the Matches

```python
Win_Matches_Count = df['winner'].value_counts()
data=[] 
teams=[]
for team in Win_Matches_Count.index:
    teams.append(team)
    data.append(Win_Matches_Count.loc[team])

bar_width = 0.4
x = range(len(teams))
plt.bar(x, data, width=bar_width, align='center')
plt.xlabel('Teams')
plt.ylabel('Frequency')
plt.title('Distribution of Won the Matches')
plt.xticks([i for i in x], teams, rotation=45, ha='right')
plt.show()
```

## 25) Ratio between Total Matches and Win Matches

```python
Total_Matches_Counts = pd.concat([df['team1'],df['team2']]).value_counts()
Total_Matches = Total_Matches_Counts.index
Win_Matches_Count = df['winner'].value_counts()
for team in Total_Matches:
    print(team+' - '+str((Total_Matches_Counts.loc[team]/Win_Matches_Count.loc[team]).round(4)))
```
